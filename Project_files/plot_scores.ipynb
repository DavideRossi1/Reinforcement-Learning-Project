```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
SARSA_noboost_noPM=pd.read_csv('scores/scores_SARSA_noboost_noPM.txt',sep=',',comment='#',header=None)
SARSA_boost_noPM=pd.read_csv('scores/scores_SARSA_boost_noPM.txt',sep=',',comment='#',header=None)
SARSA_hybrid_noPM=pd.read_csv('scores/scores_SARSA_hybrid_noPM.txt',sep=',',comment='#',header=None)
SARSA_hybrid_noPM_bigenv=pd.read_csv('scores/scores_SARSA_hybrid_noPM_bigenv.txt',sep=',',comment='#',header=None)
SARSA_boost_noPM_bigcar=pd.read_csv('scores/scores_SARSA_boost_noPM_bigcar.txt',sep=',',comment='#',header=None)
SARSA_boost_noPM_bigcar2=pd.read_csv('scores/scores_SARSA_boost_noPM_bigcar2.txt',sep=',',comment='#',header=None)
SARSA_hybrid_noPM_diff=pd.read_csv('scores/scores_SARSA_hybrid_noPM_diff.txt',sep=',',comment='#',header=None)
SARSA_hybrid_noPM_diff_bigenv=pd.read_csv('scores/scores_SARSA_hybrid_noPM_diff_bigenv.txt',sep=',',comment='#',header=None)
SARSA_hybrid_noPM_diff_bigenv2=pd.read_csv('scores/scores_SARSA_hybrid_noPM_diff_bigenv2.txt',sep=',',comment='#',header=None)
```

```python
data=SARSA_boost_noPM_bigcar2

x=np.linspace(0,len(data)-1,len(data)-1)
y=data[0][:-1]
z=data[1]

fig, ax = plt.subplots()

ax.set_xlabel("Game number")
ax.set_ylabel("Score")

plt.plot(x,y)
```

```python
data1=[]
means=[]
stds=[]
medians=[]
fig, ax = plt.subplots()

ax.set_xlabel("Game number")
ax.set_ylabel("Mean score")

for i in range(0, len(data)-1): 
    new_data_point = data[0][i]  
    data1.append(new_data_point) 
    mean = np.mean(data1)  
    std=np.std(data1)
    med=np.median(data1)
    means.append(mean)  
    stds.append(std)
    medians.append(med)
```

```python
plt.plot(x,medians)
```

```python
import seaborn as sns

sns.set()
N = len(data)-1

plt.plot(x, means, 'b-', label='means')
means=np.array(means)
stds=np.array(stds)

plt.fill_between(x, means - stds, means + stds, color='b', alpha=0.2)

plt.legend(title='title')
plt.show()
```
